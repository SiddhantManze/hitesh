```python
#prg 8 Region Filling
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import canny
from skimage import data,morphology
from skimage.color import rgb2gray
import scipy.ndimage as nd
plt.rcParams["figure.figsize"] = (12,8)
%matplotlib inline
import cv2 as cv

page_image = cv.imread('/content/lena_gray_256.tif')
#to convert rgb to gray # 3 r g b planes to single gray plane
rocket_wh=page_image.mean(axis=2)

# apply edge segmentation
# plot canny edge detection
edges = canny(rocket_wh)
plt.imshow(edges, interpolation='gaussian')
plt.title('Canny detector')
```
