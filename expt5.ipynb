```python
import cv2
import numpy as np
import matplotlib.pylab as plt
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/lena_gray_256.tif')
t = img.mean(axis=2)
```

```python
m,n = t.shape
mask = np.ones([3,3], dtype = int)
mask[2,2]= -8
```

```python
img_new = np.zeros([m,n])
# img_new = np.zeros([m,n])
```

```python
for i in range(1, m-1):
    for j in  range(1, n-1):
         temp = t[i-1, j-1]*mask[0, 0]+t[i-1, j]*mask[0, 1]+t[i-1, j + 1]*mask[0, 2]+t[i, j-1]*mask[1, 0]+ t[i, j]*mask[1, 1]+t[i, j + 1]*mask[1, 2]+t[i + 1, j-1]*mask[2, 0]+t[i + 1, j]*mask[2, 1]+t[i + 1, j + 1]*mask[2, 2]
         img_new[i, j]= temp

img_new = img_new.astype(np.uint8)
```
